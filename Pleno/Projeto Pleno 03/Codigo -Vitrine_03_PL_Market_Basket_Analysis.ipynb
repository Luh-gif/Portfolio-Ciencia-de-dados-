# Vitrine 03 (Nível Pleno): Otimização de Ticket Médio e Estratégia de Bundling
**Foco:** Complementariedade de Itens, Correlação de Demanda (Preço) e Mix de Produtos.

Neste case, assumimos o papel de Especialistas em Revenue Management. Analisamos a 'Cesta de Café da Manhã' global para identificar itens que devem ser vendidos em conjunto (combos) para maximizar o ticket médio e a margem bruta.

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')

DATA_PATH = '../data/processed/breakfast_basket_limpo.csv'
df = pd.read_csv(DATA_PATH)
df.head()
```

## 1. Benchmarking de Preço da Cesta por Região
Qual o potencial de extração de valor em cada geografia?

```
fig_region = px.box(df, x='Region', y='Breakfast_Basket_USD', color='Region',
                   title='<b>Variabilidade do Ticket Médio:</b> Custo da Cesta por Região',
                   labels={'Breakfast_Basket_USD': 'Custo Total USD'})

fig_region.update_layout(showlegend=False, template='plotly_white')
fig_region.show()
```

## 2. Matriz de Correlação: Identificando Itens Complementares
Itens com alta correlação de preço tendem a ser consumidos juntos ou possuem drivers de custo similares, tornando-os candidatos perfeitos para Bundling (venda casada).

```
pivot_df = df.pivot_table(index=['City', 'Month'], columns='Item', values='Price_USD').dropna()
corr = pivot_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='RdYlGn', fmt='.2f')
plt.title('Matriz de Correlação de Preços: Base para Estratégia de Combo')
plt.show()
```

## 3. Análise de Fator: Leite + Pão (A âncora da Cesta)
Visualizaremos se o aumento em itens básicos afeta a cesta de forma desproporcional.

```
fig_scatter = px.scatter(pivot_df, x='Milk (1 Liter)', y='Fresh White Bread (500g)', 
                      trendline='ols', opacity=0.5,
                      title='<b>Relação de Itens Âncora:</b> Milk & Bread (Combo Essencial)')
fig_scatter.show()
```

### Recomendações de Negócio (Revenue Strategy)
1. **Combo Essencial (Bread + Milk + Eggs):** Estes itens possuem correlação altíssima (>0.85). Um desconto tático no bundle pode atrair fluxo para a loja.
2. **Cross-Sell de Frutas:** Maçãs e Bananas possuem menor correlação com os básicos, podendo servir como itens de 'Upsell' para aumentar a margem da cesta.
3. **Alerta de Inflação:** Regiões com alta volatilidade em Carne e Frango devem ter combos de substituição (Proteínas Alternativas).
