# Vitrine 04 (Nível Pleno): Segmentação Oculta B2C e Definição de Personas
**Foco:** Machine Learning (Clustering), Estratégia de Customer Success e LTV.

Neste projeto, utilizamos **K-Means Clustering** para encontrar padrões não óbvios no comportamento de compra de 11k+ consumidores. O objetivo é sair de uma visão genérica e criar estratégias personalizadas de retenção e upsell.

```
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')

DATA_PATH = '../data/processed/consumer_shopping_trends_clusterizado.csv'
df = pd.read_csv(DATA_PATH)
df.head()
```

## 1. Mapa de Grupos de Consumo
Reduzimos as 25 dimensões para 2 (PCA) para visualizar a separação clara dos grupos.

```
numeric_cols = df.select_dtypes(include=['float64']).columns
pca = PCA(n_components=2)
pca_res = pca.fit_transform(df[numeric_cols])
df['pca1'] = pca_res[:, 0]
df['pca2'] = pca_res[:, 1]

fig_pca = px.scatter(df, x='pca1', y='pca2', color='Cluster', 
                     title='<b>Clusters de Consumo:</b> Visualização 2D via PCA',
                     opacity=0.7, template='plotly_dark')
fig_pca.show()
```

## 2. Caracterização das Personas
Analisamos os centros dos clusters para dar 'nome e rosto' aos grupos.

```
persona_stats = df.groupby('Cluster')[['monthly_income', 'avg_online_spend', 'impulse_buying_score', 'brand_loyalty_score']].mean()
persona_stats.index = ['Core Lovers', 'Deal Seekers', 'High Potential', 'Window Shoppers'] # Exemplo hipotético

fig_radar = px.line_polar(persona_stats.reset_index(), r='avg_online_spend', theta='index', 
                           line_close=True, title='<b>Comparativo de Gasto Online por Persona</b>')
fig_radar.show()
```

### Estratégias de Recomendação
1. **Core Lovers:** Focar em programas de fidelidade e exclusividade de marca.
2. **Deal Seekers:** Campanhas agressivas de cashback e cupons semanais.
3. **High Potential:** Oferecer frete grátis e facilidades de retorno para converter em clientes recorrentes.
