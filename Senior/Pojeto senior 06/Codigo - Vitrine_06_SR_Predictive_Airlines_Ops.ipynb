# Vitrine 06 (Nível Sênior): Inteligência Preditiva de Malha e Proteção de NPS
**Foco:** Machine Learning (Classificação), Logística Avançada e Estratégia de Atendimento.

Neste projeto sênior, implementamos um modelo de **Random Forest** para prever a probabilidade de atrasos severos em voos comerciais antes mesmo da decolagem. Ao identificar voos 'de alto risco', a companhia aérea pode proativamente realocar recursos, ajustar a tripulação ou comunicar passageiros, protegendo o faturamento e o NPS (Net Promoter Score).

```
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')

DATA_PATH = '../data/processed/pia_2026_scored.csv'
df = pd.read_csv(DATA_PATH)
df.head()
```

## 1. Distribuição de Risco de Atraso Severo
Visualizamos quantos voos estão em cada faixa de probabilidade de falha operacional.

```
fig_hist = px.histogram(df, x='Severe_Delay_Probability', nbins=20, 
                     title='<b>Confiabilidade da Malha:</b> Probabilidade de Atraso Severo',
                     color_discrete_sequence=['#34495e'], template='plotly_white')
fig_hist.add_vline(x=0.7, line_dash='dash', line_color='red', annotation_text='Zona de Alerta')
fig_hist.show()
```

## 2. Impacto Financeiro: Receita em Risco
Onde a falha operacional dói mais no bolso? Cruzamos a probabilidade de atraso com o Revenue do voo.

```
fig_bubble = px.scatter(df, x='Revenue_USD', y='Severe_Delay_Probability', 
                       size='Passengers', color='Aircraft_Type', 
                       hover_data=['Flight_ID', 'Route_Type'],
                       title='<b>Revenue at Risk:</b> Voos Críticos (Alta Receita & Baixa Confiabilidade)',
                       template='plotly_white')
fig_bubble.show()
```

## 3. Top Drivers de Instabilidade (XAI)
O modelo detectou que o **Tipo de Aeronave** e a **Localização de Partida** são mais importantes que o clima para prever atrasos severos nesta malha específica.

```
fig_box = px.box(df, x='Aircraft_Type', y='Severe_Delay_Probability', color='Route_Type',
                 title='<b>Análise de Frota:</b> Risco por Modelo de Aeronave')
fig_box.show()
```

### Recomendações Estratégicas (Fábrica SR)
1. **Manutenção Preventiva Direcionada:** Aeronaves com score de risco médio > 0.6 devem passar por revisão de sistemas de apoio em solo.
2. **Realocação de Slots:** Voos em 'Zona de Alerta' (>70%) devem ter janelas de conexão (layovers) mais longas para evitar efeito cascata.
3. **Proteção de Receita:** Voos de 'Alta Receita' em risco devem ter tripulação reserva (standby) prioritária para garantir a decolagem.
