# Vitrine Sênior: Predição de LTV via BigQuery ML
**Cenário:** E-commerce *TheLook*

Este notebook demonstra a capacidade de integrar Python com BigQuery para realizar análises de larga escala e modelagem preditiva diretamente no Data Warehouse.

```python
from google.cloud import bigquery
import pandas as pd
import plotly.express as px

client = bigquery.Client()

query = """
SELECT 
    traffic_source,
    COUNT(id) as user_count,
    AVG(age) as avg_age
FROM `bigquery-public-data.thelook_ecommerce.users` 
GROUP BY 1
"""

df = client.query(query).to_dataframe()
df.head()
```

## 1. Feature Engineering Avançada
Utilizamos CTEs para agregar o comportamento transacional dos clientes.

```python
sql_features = """
WITH user_stats AS (
  SELECT 
    user_id,
    COUNT(order_id) as total_orders,
    SUM(sale_price) as lifetime_revenue
  FROM `bigquery-public-data.thelook_ecommerce.order_items` 
  GROUP BY 1
)
SELECT * FROM user_stats LIMIT 10
"""
df_features = client.query(sql_features).to_dataframe()
df_features.head()
```
